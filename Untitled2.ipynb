## Conversion of JSON to readable dataframe


```python
response = requests.get(url, headers=headers, params=querystring)

df = pd.DataFrame(response.json()['data']['data'])[['name', 'hasMenu', 'menuUrl', 'establishmentTypeAndCuisineTags', 'reviewSnippets','locationId','restaurantsId', ]]
```

# Feature Conversion

```python
df['establishmentTypeAndCuisineTags'] = df['establishmentTypeAndCuisineTags'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))
```
