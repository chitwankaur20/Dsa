```python
url = "https://timesofindia.indiatimes.com/"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
r = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
articles = soup.select("div.nEjlO")
list_art =[]
```

```python

for div in articles:
    # Extract all text from child tags (like <a>, <span>, etc.) if available
    for child in div.find_all(recursive=True):  # Gets all nested tags
        text = child.get_text(strip=True)
        if text and len(text) > 10:  # Filter out empty/short junk text
            list_art.append(text)
```

```python
import re
import pandas as pd

# Pattern: split where lowercase/punctuation is followed by uppercase (no space)
split_pattern = r'(?<=[a-z0-9\?\.\!\'"])(?=[A-Z])'

# Apply the split
headlines = re.split(split_pattern, text)

# Clean headlines
headlines = [h.strip().strip('\'"') for h in headlines if len(h.strip()) > 10]

# Remove noisy tokens
noise = ['Subscribe', 'TOI+', 'Exclusive Stories']
headlines = [h for h in headlines if not any(n in h for n in noise)]

# Convert to DataFrame
df = pd.DataFrame(headlines, columns=["headline"])

# Print first 15 rows
print(df.head(15))
```
